```python
import matplotlib.pyplot as plt
import seaborn as sns

# Columns ki list
columns_list = ['bmi', 'avg_glucose_level', 'age']  # Apne actual column names yahan replace karein

# Boxplot banane ka loop
plt.figure(figsize=(10, 7))  # Graph ka size adjust karein
for column in columns_list:
    plt.figure()  # Har column ke liye ek nayi figure
    sns.boxplot(data=data, x=column)  # df ko apne DataFrame se replace karein
    plt.title(f'Boxplot for {column}')  # Title add karein
    plt.show()  # Graph show karein
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data['gender'] = label_encoder.fit_transform(data['gender'])
data['ever_married'] = label_encoder.fit_transform(data['ever_married'])
data['work_type'] = label_encoder.fit_transform(data['work_type'])
data['Residence_type'] = label_encoder.fit_transform(data['Residence_type'])
data['smoking_status'] = label_encoder.fit_transform(data['smoking_status'])
```

```python
data['smoking_status'].value_counts()
```

```python
data
```

```python
X = data.drop(columns=['stroke'],axis=1)
```

```python
y=data.stroke
```

```python
#stroke class is imbalance we see in up so now balance it 
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
# Step 2: Ab check karein balanced data
print(y_resampled.value_counts()) 
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Train a Random Forest model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
# The trained model now has a feature_importances_ attribute
importances = model.feature_importances_
```

```python
# Create a DataFrame mapping each feature to its importance
importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': importances
})

# Sort the DataFrame for better visualization
importance_df = importance_df.sort_values(by='Importance', ascending=True)

# Plot the horizontal bar chart
plt.figure(figsize=(10, 8))
plt.barh(importance_df['Feature'], importance_df['Importance'], color='skyblue')
plt.xlabel('Feature Importance')
plt.title('Feature Importance Graph')
plt.show()
```

```python
from sklearn.metrics import f1_score
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print("F1-Score:", f1_score(y_test, y_pred))
```

```python
print(y_pred)
print("---------------------*--------------------")
print(y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy : {accuracy}")
```
